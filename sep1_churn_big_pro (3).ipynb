```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('Telco-Customer-Churn.csv')
```

```python
y=dataset['Churn']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
y=le.fit_transform(y.values.ravel())
```

```python
y=pd.DataFrame(y)
```

```python
y
```

```python
dataset=dataset.drop(['customerID','Churn'],axis=1)
```

```python
dataset
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocess=ColumnTransformer([('sc',StandardScaler(),['tenure','MonthlyCharges','TotalCharges']),('encoder',OneHotEncoder(sparse_output=False),['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod'])])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
pipeline_logistic=Pipeline([('model_logis',LogisticRegression(random_state=42,solver='liblinear',C=0.01,penalty='l2'))])
```

```python
x=dataset
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train['TotalCharges'] = x_train['TotalCharges'].replace(' ', np.nan)
nan_indices_train = x_train[x_train['TotalCharges'].isna()].index
x_train.dropna(inplace=True)
y_train = y_train.drop(nan_indices_train)
x_train=preprocess.fit_transform(x_train)
```

```python
pipeline_logistic.fit(x_train,y_train.values.ravel())
```

```python
x_test['TotalCharges'] = x_test['TotalCharges'].replace(' ', np.nan)
nan_indices_test = x_test[x_test['TotalCharges'].isna()].index
x_test.dropna(inplace=True)
y_test = y_test.drop(nan_indices_test)
x_test=preprocess.fit_transform(x_test)
```

```python
y_pred_logistic=pipeline_logistic.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc_logistic=accuracy_score(y_test,y_pred_logistic)
```

```python
acc_logistic
```

```python
print(confusion_matrix(y_test,y_pred_logistic))
```

```python
print(classification_report(y_test,y_pred_logistic))
```
